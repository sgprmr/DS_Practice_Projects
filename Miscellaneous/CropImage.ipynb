```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from scipy import ndimage, misc
```

```python
pic = mpimg.imread("./Test_pic.PNG")  
```

```python
pic.shape
```

```python
def crop(pic,x1,x2,y1,y2):
    return pic[y1:y2,x1:x2]
```

```python
cropImage = crop(pic,0,245,0,254)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(8, 4))
ax = axes.ravel()

ax[0].imshow(pic)
ax[0].set_title("Original")
ax[1].imshow(cropImage, cmap=plt.cm.gray)
ax[1].set_title("cropImage")

fig.tight_layout()
plt.show()
```
